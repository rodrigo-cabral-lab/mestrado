```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df = pd.read_excel('particle_tracking.xlsx', sheet_name=['45g','63g','72g'])
```

```python
df_45 = df['45g']
df_63 = df['63g']
df_72 = df['72g']
```

```python
def plotparticle(distance,particle):
    
    for (d) in (distances):
        x = df_45['x_'+particle+'_d=1e-3_'+d] 
        y = df_45['y_'+particle+'_d=1e-3_'+d]
        plt.scatter(x,y, label='Diameter')
        plt.xlim(0.15,0.3)
        #plt.ylim(-0.5,2.5)
        plt.grid(True)
        plt.legend(loc="best", prop={'size': 20})   
        plt.show;
```

```python
distances=['5d','tailedge']
#size=[3]
plotparticle(distances,'co2')
```

```python
def plotparticle45(particle):
    plt.figure(figsize=(12,10))
    for p in (particle):
        x1 = df_45['x_'+p+'_d=1e-3_inlet'] 
        y1 = df_45['y_'+p+'_d=1e-3_inlet']
        plt.scatter(x1,y1, label=p+' $d=1e^{-3}$')
        x2 = df_45['x_'+p+'_d=1e-4_inlet'] 
        y2 = df_45['y_'+p+'_d=1e-4_inlet']
        plt.scatter(x2,y2, label=p+' $d=1e^{-4}$')
        plt.xlim(0,0.2)
        #plt.ylim(-0.5,2.5)
        plt.grid(True)
        plt.legend(loc="best", prop={'size': 20})   
        plt.show;
```

```python
particle=['co2','h2s','ch4']
plotparticle45(particle)
plt.title('45',  fontdict={'size': 16});
```
