```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df = pd.read_excel('particle_tracking.xlsx', sheet_name=['45g','63g','72g'])
```

```python
df_45 = df['45g']
df_63 = df['63g']
df_72 = df['72g']
```

# Case 45ยบ

```python
def plotparticle45(particle, position):
    plt.figure(figsize=(16,10))
    for x in position:
        for p in (particle):
            x1 = df_45['x_'+p+'_d=1e-3_'+x] 
            y1 = df_45['y_'+p+'_d=1e-3_'+x]
            plt.scatter(x1,y1, label=p+', $d=1\cdot10^{-3}$, '+x)
            x2 = df_45['x_'+p+'_d=1e-4_'+x]
            y2 = df_45['y_'+p+'_d=1e-4_'+x]
            plt.scatter(x2,y2, label=p+', $d=1\cdot10^{-4}$, '+x)
            plt.xlim(0,0.4)
            #plt.ylim(-0.5,2.5)
            plt.xlabel('x [m]', fontdict={'size': 20})
            plt.ylabel('r [m]', fontdict={'size': 20})
            plt.rcParams['xtick.labelsize'] = 20
            plt.rcParams['ytick.labelsize'] = 20
            plt.grid(True)
            plt.legend(loc="best", prop={'size': 20})   
            plt.show;
```

```python
particle=['h2s']
# particle: inlet, 5d, tailedge
x=['inlet','5d','tailedge']
#for x in position:
plotparticle45(particle, x)
plt.title('45ยบ',  fontdict={'size': 20});
```
