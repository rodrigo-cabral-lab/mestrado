```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

## Validação laminar (Ayinde)

```python
df = pd.read_excel('dados_ayinde.xlsx', sheet_name=['ayinde','numerico','s_ayinde','s_numerico'])
 
```

```python
def plotVel(distances, direction):
  #distances = [0,10,20,40,80]
  plt.figure(figsize=(10, 6))
  for d in distances:
    plt.plot(df['ayinde']['r_vel' + direction  + '_' + str(int(d)) ], df['ayinde']['vel_' + direction  + '_' + str(int(d)) ], label=str(int(d))+'D - Ayinde')
    plt.scatter(df['numerico']['r_' + str(int(d)) ], df['numerico']['vel_' + direction  + '_' + str(int(d)) ], label=str(int(d))+'D - Númerico')

  plt.xlabel("r/R")
  if direction == 'tang':
    plt.ylabel("v/Umax")
  elif direction == 'axial':
    plt.ylabel("u/Umax")
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
distances = [0,10,20,40,80]
plotVel(distances, 'axial')
plt.savefig('graficos/vel_axial_ayinde.png')
plotVel(distances, 'tang')
plt.savefig('graficos/vel_tang_ayinde.png')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['s_ayinde']['x_re1000'], df['s_ayinde']['s_re1000'], label= 'Re 1000 - Ayinde')
plt.plot(df['s_ayinde']['x_re1800'], df['s_ayinde']['s_re1800'], label= 'Re 1800 - Ayinde')
plt.scatter(df['s_numerico']['x'], df['s_numerico']['s_re1000'], label=' Re 1000 - Numérico')
plt.scatter(df['s_numerico']['x'], df['s_numerico']['s_re1800'], label=' Re 1800 - Numérico')

plt.xlabel("x/D")
plt.ylabel("S/S0")
plt.grid(True)
plt.legend(loc="best")   
plt.show;
plt.savefig('graficos/s_validacao_ayinde.png')
```
