```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

## Turbulent validation (Dirkzwager)

```python
df = pd.read_excel('dados_dirkz.xlsx', sheet_name=['dirkz','numerico'])
```

## Comparison of axial and tangential velocity profiles (numerical and literature)

```python
def plotVel(distances, direction):
  #distances = [5,10,15,20]
  #plt.figure(figsize=(10, 6))
  plt.figure(figsize=(10, 6))
  for d,i in zip(distances,plot):
   plt.subplot(2,2,i)
   plt.scatter(df['dirkz']['r_vel' + direction  + '_' + str(int(d)) ], df['dirkz']['vel_' + direction  + '_' + str(int(d)) ], label=str(int(d))+'D - Dirkzwager')
   plt.scatter(df['numerico']['r_' + str(int(d)) ], df['numerico']['vel_' + direction  + '_' + str(int(d)) ], label=str(int(d))+'D - Numerical')
 
   plt.xlabel("r [m]")
   if direction == 'tang':
    plt.ylabel("v [m/s]")
   elif direction == 'axial':
    plt.ylabel("u [m/s]")
   plt.grid(True)
   plt.legend(loc="best")   
   plt.show;
 
```

```python
distances = [5,10,15,20]
plot = [1,2,3,4]
plotVel(distances, 'axial')
plt.savefig('graficos/vel_axial_ayinde.png')
plotVel(distances, 'tang')
plt.savefig('graficos/vel_tang_ayinde.png')
```
