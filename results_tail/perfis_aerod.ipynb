```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_excel('perfis_power_series.xlsx', sheet_name=['perfis','cp','tau'])
```

```python
def plotPerfis2(cases):
  plt.figure(figsize=(10, 6))
  for c in cases:
    plt.plot(df['perfis']['x_caso'+ str(int(c))], df['perfis']['y_caso'+ str(int(c))], label='Case '+str(int(c)))
    
  
  plt.plot(40,11.5,'rs') 
  plt.plot(51,12,'rs') 
  plt.plot(105,11.7,'rs') 
  plt.plot(31,11.6,'rs')
  plt.plot(40,11.9,'rs') 
  plt.plot(79,12,'rs') 
    
  plt.xlabel("x [mm]")
  plt.ylabel("P(x) [mm]")
  plt.title('Perfis Power Series')
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
cases = [1,2,3,4,5,6]
plotPerfis2(cases)
plt.savefig('graficos/perfis_powerseries_2.png')
```

```python
def plotPerfis(cases):
  plt.figure(figsize=(10, 6))
  for c in cases:
    plt.plot(df['perfis']['x_caso'+ str(int(c))], df['perfis']['y_caso'+ str(int(c))], label='Case '+str(int(c)))

    
  plt.xlabel("x [mm]")
  plt.ylabel("P(x) [mm]")
  #plt.title('Perfis Power Series')
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
cases = [1,2,3,4,5,6]
plotPerfis(cases)
plt.savefig('graficos/perfis_powerseries.png')
```

```python
def plotCp(cases):
  plt.figure(figsize=(10, 6))
  for c in cases:
    plt.plot(df['cp']['x_caso'+ str(int(c))], df['cp']['y_caso'+ str(int(c))], label='Case '+str(int(c)))
    
  plt.xlabel("Curve length [m]")
  plt.ylabel("Pressure coefficient")
  plt.ylim(0, 5.5)
  plt.xlim(0, 0.10)
  #plt.title('Perfis Power Series')
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
cases = [1,2,3,4,5,6]
plotCp(cases)
plt.savefig('graficos/coef_pressao_perfis.png')
```

```python
def plotTau(cases):
  plt.figure(figsize=(10, 6))
  for c in cases:
    plt.plot(df['tau']['x_caso'+ str(int(c))], df['tau']['y_caso'+ str(int(c))], label='Case '+str(int(c)))
    
  plt.xlabel("Curve length [m]")
  plt.ylabel("Shear stress [Pa]")
  #plt.title('Perfis Power Series')
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
cases = [1,2,3,4,5,6]
plotTau(cases)
plt.savefig('graficos/tensao_cisalhamento_perfis.png')
```
