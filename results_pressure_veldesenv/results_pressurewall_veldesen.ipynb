```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_excel('results_pressure_veldens.xlsx', sheet_name=['pressure_wall','velaxial_developed'])
```

```python
def plotPressurewall():
  plt.figure(figsize=(10, 6))
  plt.scatter(df['pressure_wall']['z_d_exp'], df['pressure_wall']['pressure_wall_exp'], label='Dirkzwager')
  plt.scatter(df['pressure_wall']['z_d_num'], df['pressure_wall']['pressure_wall_num'], label='Numerical')
  plt.xlabel("z/D")
  plt.ylabel("Pressure Static [Pa]")
  #plt.title('Pressure wall')
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
plotPressurewall()
plt.savefig('graphics/pressure_wall.png')
```

```python
def plotVeldev(section):
  plt.figure(figsize=(10, 6))
  plt.plot(df['velaxial_developed']['z_d'], df['velaxial_developed']['velaxial_theoretical'], label='Theoretical profile')
  for s in section:
    plt.scatter(df['velaxial_developed']['z_d'], df['velaxial_developed']['velaxial_num_'+str(int(s))+'d'], label='Profile '+str(int(s))+'D')
    
  plt.xlabel("z/D")
  plt.ylabel("u [m/s]")
  #plt.title('Profiles Vel axial')
  plt.grid(True)
  plt.legend(loc="best")   
  plt.show;
```

```python
section=[100,110,120,130,135]
plotVeldev(section)
plt.savefig('graphics/vel_developed.png')
```
